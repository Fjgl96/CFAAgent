```python
INDEX_NAME = "cfa_documents"
```

```python
print(f"Conectando a {ES_URL}...")

# Ignorar advertencias si usamos verify_certs=False (aunque usaremos True)
warnings.filterwarnings('ignore', 'Unverified HTTPS request')

try:
    # 1. Conectar (usando la configuraci√≥n que funcion√≥)
    client = Elasticsearch(
        [ES_URL],
        basic_auth=(ES_USER, ES_PASS),
        verify_certs=True,
        request_timeout=15
    )

    if not client.ping():
        print("‚ùå ERROR: No se pudo hacer ping. Revisa tus credenciales.")
    else:
        print("‚úÖ Conexi√≥n exitosa.")
        
        # 2. Verificar si el √≠ndice existe ANTES de borrar
        print(f"Verificando si el √≠ndice '{INDEX_NAME}' existe...")
        
        if client.indices.exists(index=INDEX_NAME):
            print(f"üóëÔ∏è El √≠ndice existe. Borrando '{INDEX_NAME}'...")
            
            # 3. Borrar el √≠ndice
            response = client.indices.delete(index=INDEX_NAME, ignore=[400, 404])
            
            if 'acknowledged' in response and response['acknowledged']:
                print(f"‚úÖ ¬°√âXITO! √çndice '{INDEX_NAME}' borrado.")
            else:
                print(f"‚ö†Ô∏è El borrado no fue reconocido: {response}")
                
        else:
            print(f"‚ÑπÔ∏è El √≠ndice '{INDEX_NAME}' no existe. No hay nada que borrar.")

except Exception as e:
    print(f"‚ùå Ocurri√≥ un error inesperado: {e}")

print("\n--- Proceso finalizado ---")
```
